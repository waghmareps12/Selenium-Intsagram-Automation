```python
import pandas as pd
import json
```

```python
# Read the JSON file
with open(r'C:\Users\waghm\Downloads\pending_follow_requests (2).json', 'r') as file:
    json_data = json.load(file)
```

```python
# Extract the relevant fields
rows = []
for item in json_data['relationships_follow_requests_sent']:
    for string_data in item['string_list_data']:
        rows.append(string_data)

# Create a DataFrame
df = pd.DataFrame(rows)
```

```python
df = df.sort_values(by = 'date')
```

```python
import json
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
from selenium.webdriver.chrome.options import Options
```

```python
options = Options()
options.add_experimental_option("debuggerAddress", "localhost:9292")
driver = webdriver.Chrome(options=options)
```

```python
df = df[df.index<2142]
```

```python
# Function to log into Instagram
def login_to_instagram(username, password):
    driver.get("https://www.instagram.com")
    time.sleep(5)

    username_input = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.NAME,  "Phone number, username, or email" ))
    )
    password_input = driver.find_element(By.NAME,  "Password" )

    username_input.send_keys(username)
    password_input.send_keys(password)
    password_input.send_keys(Keys.RETURN)

    time.sleep(5)    

# Function to unfollow a user
def unfollow_user(profile_url):
    driver.get(profile_url)
    time.sleep(1)

    unfollow_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, "//div[@class='_ap3a _aaco _aacw _aad6 _aade']"))
    )
    unfollow_button.click()
    time.sleep(1)

    confirm_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, "//button[@class='_a9-- _ap36 _a9-_']"))
    )
    confirm_button.click()
    time.sleep(2)
```

```python

# Setup Selenium WebDriverptions = webdriver.ChromeOptions()
# options.add_argument("--start-maximized")

count = 180

# # Log into Instagram (replace 'your_username' and 'your_password' with your credentials)


# Loop through each URL and unfollow the user
for url in df['href']:
    
        try:
            unfollow_user(url)
            print(f"Unfollowed: {url} {count}")
            count+=1

        except Exception as e:
            print(f"Failed to unfollow {url}: {e}")

# lose the browser
driver.quit()
```
